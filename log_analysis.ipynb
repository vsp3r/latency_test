```python
import pandas as pd
```

```python
log_df = pd.read_csv('hl.log')[40:]
log_df['exch_to_server'] = log_df['exch_to_server'] / 1000
```

```python
print(log_df.describe())
```

```python
import matplotlib.pyplot as plt

# Plotting exchange to server times
plt.figure(figsize=(10, 6))
plt.plot(log_df.index, log_df['exch_to_server'], label='Exchange to Server Time')
plt.xlabel('Timestamp')
plt.ylabel('Time (ms)')
plt.title('Exchange to Server Time Over Time')
plt.legend()
plt.show()

# Plotting message to process times
plt.figure(figsize=(10, 6))
plt.plot(log_df.index, log_df['msg_to_process'], label='Message to Process Time')
plt.xlabel('Timestamp')
plt.ylabel('Time (us)')
plt.title('Message to Process Time Over Time')
plt.legend()
plt.show()
```
